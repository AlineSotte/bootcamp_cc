#### Exercício 1 - No dicionário 'my_dict' são armazenados as notas de alunos
##### Imprima a nota do aluno 'Júlio'


```python
my_dict = {
    'Ana': 10,
    'Júlio': 8,
    'Antônio': 9,
    'Maria': 7
}

for pessoa,nota in my_dict.items():
   if pessoa == 'Júlio':
       print(nota)
```
