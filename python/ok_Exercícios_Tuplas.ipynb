##### Exercício 1 - Calcule a soma dos elementos da  tupla 'my_tuple'


```python
my_tuple = 1, 2, 3, 4, 5
soma=sum(my_tuple)
print(f'soma da tupla é {soma}')
```

##### Exercício 2 - Retorne o segundo elemento da tupla 'my_tuple'


```python
my_tuple = 1, 2, 3, 4, 5
segundo_termo = my_tuple[1]
print(segundo_termo)
```

##### Exercício 3 - Encontrar o erro no código abaixo

```python
my_tuple = 1, 2, 3, 4, 5
my_tuple.remove(1) # não existe essa função
print(my_tuple)
```

```python
#correção
item_para_remover = 2
nova_tupla = tuple(item for item in my_tuple if item != item_para_remover)
print(nova_tupla)
```
