Um **loop for** é uma instrução para iterar sobre uma sequência (lista, dicionário, entre outros), permitindo que o código seja executado várias vezes.

for.png


*Exercício 1: Impressão de Números Pares*

Crie um loop for que imprime todos os números pares de 1 a 10.

```python
numero=1
for numero in range(1, 11):
    if numero % 2==0:
        print(numero)
```

*Exercício 2: Soma de Números Ímpares*

Crie um loop for que calcula a soma de todos os números ímpares de 1 a 20 e imprime o resultado.

```python
soma=0
for numero in range(1,21):
    if numero % 2!=0:
        soma=soma+numero
print(soma)        
```

*Exercício 3: Impressão de Nomes*

Crie uma lista de nomes de frutas: "maçã", "banana", "laranja", "uva", "morango".
Use um loop for para imprimir cada nome de fruta da lista.

```python
frutas =["maçã", "banana", "laranja", "uva", "morango"]
for fruta in frutas:
    print(fruta)
```

*Exercício 4: Contagem de Vogais*

Crie uma string contendo uma frase de sua escolha.
Use um loop for para contar e imprimir o número de vogais (a, e, i, o, u) na frase.

```python
frase= 'O mundo e um moinho'

contador= 0

for letra in frase.lower():
    if letra in "aeiou":
        contador += 1

print("Número de vogais:", contador)
```

*Exercício 5: Tabuada de Multiplicação*

Crie um loop for que calcula e imprime a tabuada de multiplicação de um número inteiro fornecido pelo usuário.

```python
numero = int(input('digite um numero inteiro:'))
for tabuada in range(1,11):
    print(tabuada*numero)
```
