#### Exercício 1
Apresentar o quadrado dos números inteiros de 15 a 200

```python
numero = 15
while numero <= 200:
    numero_quadro = numero*numero
    numero += 1
    print(numero_quadro)
print("Loop concluído!")
```

#### Exercício 2

Apresentar o total da soma obtida dos cem primeiros números inteiros
(1+2+3+4+5+6+7+...+97+98+99+100)

```python
numero = 1 
soma =0
while numero <= 100:
    soma = soma + numero
    numero= numero+1
print(soma)
```

#### Exercício 3
Escrever um algoritmo que lê 5 valores para a variável A, um de cada vez,
e contar quantos destes valores são negativos. Após a leitura, o programa deve mostrar
a quantidade de números negativos

```python
A = [-10,2,5,-2,1]

contador = 0
negativos = 0

while contador < 5:
    if A[contador] < 0:
        negativos += 1
    contador += 1

print("Quantidade de números negativos:", negativos)
```

#### Exercício 4
Ler duas notas e informar a média. Se o valor digitado for menor do que 0 (zero)
ou maior do que 10 (dez), o usuário deve digitar a nota novamente

```python
contador = 0
soma = 0

while contador < 2:
    nota = float(input("Digite uma nota (0 a 10): "))

    while nota < 0 or nota > 10:
        print("Nota inválida! Digite novamente.")
        nota = float(input("Digite uma nota (0 a 10): "))

    soma += nota
    contador += 1

media = soma / 2
print("Média:", media)
```
