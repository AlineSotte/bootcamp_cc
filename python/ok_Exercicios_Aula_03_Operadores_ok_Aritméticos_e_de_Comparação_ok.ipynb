#### **Operadores Aritméticos**

Os operadores aritméticos são usados ​​com valores numéricos para realizar operações matemáticas.

Operador | Nome
-------- | --------
+        | Adição
-        | Subtração
*        | Multiplicação
**       | Exponenciação
/        | Divisão
//       | Divisão Inteira (Quociente)
%        | Módulo (Resto da Divisão)

#### **Operadores de Comparação**

Os operadores de comparação são usados ​​para comparar dois valores.

Operador  | Nome
--------  | --------
==        | Igual a
>         | Maior que
<         | Menor que
>=        | Maior igual
<=        | Menor igual
!=        | Diferente de

*Exercício 1: Conversão de Tipos*

*Exercício 3: Operadores Aritméticos*

Crie um programa que solicite ao usuário dois números inteiros e, em seguida, realize as seguintes operações:
- Some os números.
- Subtraia o segundo número do primeiro.
- Multiplique os números.


```python
entrada_usuario_1 = input("Digite um número inteiro: ")
entrada_usuario_2 = input("Digite um número inteiro: ")
soma = int(entrada_usuario_1)+int(entrada_usuario_2)
subtracao = int(entrada_usuario_2) - int(entrada_usuario_1)
resultado_multiplicacao = soma * subtracao
print(f'Resultado {resultado_multiplicacao}')
```

*Exercício 4: Operadores de Comparação Lógica com Número de Referência*

Crie um programa que solicite ao usuário um número inteiro.
Peça ao usuário para inserir um número de referência.
Em seguida, verifique e imprima se o número inserido pelo usuário é:
- Igual ao número de referência.
- Maior que o número de referência.
- Menor que o número de referência.

```python
entrada = int(input("Digite um número inteiro: "))
entrada_referencia=int(input("Digite um número inteiro de referencia: "))

if entrada == entrada_referencia:
    print('Igual ao número de referência')
elif entrada > entrada_referencia:
    print('Maior que o número de referência')
else :
    print( 'Menor que o número de referência')
```
