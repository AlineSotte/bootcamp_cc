<br>
<br>

___
As estruturas de controle servem para decidir quais blocos de código serão executados.


Exemplo:

    Se estiver nublado:
  
        Levarei guarda-chuva

    Senão:
      
        Não levarei


No Python, os comandos de controle de fluxo é o **if** e o **else**.

```
if <condicao>:
  comando
```

*Exercício 1: Verificação de Paridade*

Crie um programa que solicita ao usuário um número inteiro e, em seguida, imprime se o número é par ou ímpar.

````python
entrada_usuario= int(input("Digite um número inteiro: "))
if (entrada_usuario % 2 == 0):
    print(f'O número é par')
else:
    print(f'O número é ímpar')
````

*Exercício 2: Verificação de Maioridade*

Crie um programa que solicita ao usuário sua idade e, em seguida, verifica se a pessoa é maior de idade (idade maior ou igual a 18 anos).

````python
entrada_usuario= int(input("Digite sua idade: "))
if (entrada_usuario >= 18):
    print(f'maior de idade')
else:
    print(f'menor de idade')
````

*Exercício 3: Verificação de Número Positivo, Negativo ou Zero*

Crie um programa que solicita ao usuário um número e, em seguida, verifica se o número é positivo, negativo ou igual a zero.

````python
entrada_usuario= int(input("Digite número: "))
if (entrada_usuario > 0):
    print(f'número positivo')
elif(entrada_usuario == 0):
    print(f'igual a zero')
else:
    print(f'número negativo')
````

*Exercício 4: Verificação de Número Maior*

Crie um programa que solicita ao usuário dois números inteiros e, em seguida, verifica e imprime qual deles é o maior.

````python
entrada_1= int(input("Digite o primeiro número: "))
entrada_2 = int(input("Digite o segundo número: "))

if (entrada_1 > entrada_2):
    print(f'o numero maior é {entrada_1}')
else:
    print(f'o numero maior é {entrada_2}')
````
