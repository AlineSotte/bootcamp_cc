# **Aula 07 - Estruturas de Dados – Tuplas**

<br>
<br>

___
Tupla é um tipo de estrutura de dados utilizada em Python que funciona de modo semelhante a uma lista, entretanto, com a característica principal de ser imutável.

Uma tupla consiste em uma sequência de valores separados por vírgulas.

*Exercício 1: Criação e Acesso a Elementos de uma Tupla*

Crie uma tupla chamada meses com os nomes dos meses do ano.
Acesse e imprima o terceiro mês da tupla.

```python
meses_do_ano = (
    "Janeiro", "Fevereiro", "Março", "Abril",
    "Maio", "Junho", "Julho", "Agosto",
    "Setembro", "Outubro", "Novembro", "Dezembro"
)
print(meses_do_ano[2])
```

*Exercício 2: Criação e Acesso a Elementos de uma Tupla*

Crie uma tupla com o nome e a idade de uma pessoa e escreva na tela uma saudação com as informações contidas na tupla

```python
pessoas = (
    ("Aline", 40),
    ("Pedro", 25),
    ("Rafaela", 28),
    ("Matheus", 32)
)

for nome, idade in pessoas:
    print(f"Olásss, {nome}! Você tem {idade} anos.")
```
