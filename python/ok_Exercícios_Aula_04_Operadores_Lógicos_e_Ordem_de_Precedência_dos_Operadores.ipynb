*Exercício 1: Verificação de Divisibilidade*

Crie um programa que solicita ao usuário um número e, em seguida, imprime mensagens indicando:
- Se o número é divisível por 3.
- Se o número é divisível por 4.
- Se o número é divisível por 3 e por 4.
- Se o número é divisível por 3 ou por 4.

```python
entrada_usuario= int(input("Digite um número inteiro: "))

if (entrada_usuario % 3 == 0):
    print(f'Divisivel {entrada_usuario} por 3')
if (entrada_usuario % 4 == 0):
    print(f'Divisivel {entrada_usuario} por 4')
if (entrada_usuario % 3 == 0 and entrada_usuario % 4 == 0):
    print(f'Divisivel {entrada_usuario} por 3 e 4')
if (entrada_usuario % 3 == 0 or entrada_usuario % 4 == 0):
    print(f'Divisivel {entrada_usuario} por 3 ou 4')
```
