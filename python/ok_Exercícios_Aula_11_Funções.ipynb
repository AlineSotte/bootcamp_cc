*Exercício 1: Função de Soma*

Crie uma função chamada soma que aceita dois argumentos e retorna a soma deles.
Teste a função com diferentes pares de números.

```python
def soma(a,b):
    soma=a+b
    return soma

soma1=soma(3,2)
soma2=soma(4,5)
soma3=soma(1,0)
soma4=soma(-9,2)

print(soma1,soma2,soma3,soma4)
```

*Exercício 2: Função de Média*

Crie uma função chamada media que aceita uma lista de números como argumento e retorna a média dos números.
Teste a função com diferentes listas de números.

```python
def media(lista):
    soma=0
    for numero in lista:
        soma=soma+numero
    media_numero= soma/len(lista)
    return media_numero

numeros = [1, 2, 3, 4, 5]
media(numeros)
```

*Exercício 3: Função de Verificação de Paridade*

Crie uma função chamada par_ou_impar que aceita um número como argumento e retorna "Par" se o número for par ou "Ímpar" se for ímpar.
Teste a função com diferentes números.

```python
def par_ou_impar(numero):
    if numero % 2 == 0:
        return "Par"
    else:
        return "Ímpar"

print(par_ou_impar(2))    
print(par_ou_impar(7))    
print(par_ou_impar(0))    
print(par_ou_impar(15))  
print(par_ou_impar(-4))   
```

*Exercício 4: Função de Verificação de Palíndromo*

Crie uma função chamada e_palindromo que aceita uma string como argumento e retorna True se a string for um palíndromo (uma palavra que é a mesma quando lida de trás para frente) ou False caso contrário.
Teste a função com diferentes palavras.

```python
def e_palindromo(palavra):
    palavra = palavra.lower()  
    return palavra == palavra[::-1]


print(e_palindromo("radar"))     
print(e_palindromo("arara"))     
print(e_palindromo("Python"))    
print(e_palindromo("Level"))     
print(e_palindromo("Casa")) 
```

*Exercício 5: Função de Cálculo de Fatorial*

Crie uma função chamada fatorial que aceita um número inteiro como argumento e retorna o fatorial desse número.
Teste a função com diferentes números inteiros.

```python
def fatorial(n):
    if n == 0 or n == 1:   
        return 1
    else:
        return n * fatorial(n - 1)

print(fatorial(0))  
print(fatorial(1))   
print(fatorial(2))   
print(fatorial(3))   
```
