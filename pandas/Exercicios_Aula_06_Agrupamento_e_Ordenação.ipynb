*Exercício 1: Utilizando a função groupby*

- Importe a biblioteca pandas.
- Crie um DataFrame chamado "dados_vendas" com as colunas "Produto", "Vendedor" e "Valor".
- Preencha o DataFrame com dados fictícios de vendas.
- Use a função groupby para agrupar as vendas por vendedor e calcular o total de vendas de cada vendedor.
- Imprima o resultado.

```python
import pandas as pd 
```

```python
dados_vendas = {
    "Produto": ["Notebook", "Mouse", "Notebook", "Teclado", "Mouse", "Teclado"],
    "Vendedor": ["Ana", "Bruno", "Ana", "Carla", "Bruno", "Ana"],
    "Valor": [3000, 150, 3200, 200, 180, 250]
}


df_dados_vendas = pd.DataFrame(dados_vendas)
```

```python
total_por_vendedor = df_dados_vendas.groupby("Vendedor")["Valor"].sum()
print(total_por_vendedor)
```

*Exercício 2: Utilizando a função sort_values para Classificar os Dados*

- Utilize o DataFrame "dados_vendas" criado no exercício 1.
- Use a função sort_values para classificar as vendas em ordem decrescente com base no valor.
- Imprima o DataFrame após a classificação.

```python
ordenado_valor = df_dados_vendas.sort_values(by="Valor", ascending=False)
print(ordenado_valor)
```

*Exercício 3: Utilizando a função groupby com Classificação*

- Utilize o DataFrame "dados_vendas" criado no exercício 1.
- Use a função groupby para agrupar as vendas por produto e, em seguida, use a função sort_values para classificar as vendas de cada produto em ordem crescente com base no valor.
- Imprima o resultado.

```python
produto_ordenado = df_dados_vendas.groupby("Produto", group_keys=False).apply(lambda x: x.sort_values(by="Valor"))
print(produto_ordenado)
```

*Exercício 4: Utilizando a função groupby com Classificação Personalizada*

- Utilize o DataFrame "dados_vendas" criado no exercício 1.
- Use a função groupby para agrupar as vendas por vendedor e, em seguida, use a função sort_values para classificar as vendas de cada vendedor em ordem decrescente com base no valor.
- Imprima o resultado.

```python
vendedor_ordenado = df_dados_vendas.groupby("Vendedor", group_keys=False).apply(lambda x: x.sort_values(by="Valor", ascending=False))
print(vendedor_ordenado)
```
