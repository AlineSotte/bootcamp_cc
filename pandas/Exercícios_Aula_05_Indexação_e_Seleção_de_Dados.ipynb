*Exercício 1: Utilizando a função loc para Selecionar Linhas por Rótulo*

- Importe a biblioteca pandas.
- Crie um DataFrame chamado "dados_alunos" com as colunas "Nome" e "Idade".
- Configure os índices para as letras "A", "B", "C" e "D".
- Utilize a função loc para selecionar e imprimir as informações do aluno com índice "B".

```python
import pandas as pd 
```

```python
dados_alunos = {
        "Nome": ["Ana", "Bruno", "Carla", "Daniel"],
        "Idade": [20, 22, 19, 21]}

df_dados_alunos = pd.DataFrame( dados_alunos, index=["A","B","C","D"])
df_dados_alunos.head(4)
```

*Exercício 2: Utilizando a função iloc para Selecionar Linhas por Índice Numérico*

- Utilize o DataFrame "dados_alunos" criado no exercício 1.
- Use a função iloc para selecionar e imprimir as informações do aluno na segunda linha.

```python
print(df_dados_alunos.loc["B"])
```

*Exercício 3: Utilizando a função loc para Selecionar Colunas por Rótulo*

- Utilize o DataFrame "dados_alunos" criado no exercício 1.
- Use a função loc para selecionar e imprimir a coluna "Nome" de todos os alunos.

```python
print(df_dados_alunos.loc[:, "Nome"])
```

*Exercício 4: Utilizando a função iloc para Selecionar Colunas por Índice Numérico*

- Utilize o DataFrame "dados_alunos" criado no exercício 1.
- Use a função iloc para selecionar e imprimir a coluna de índice 1 (ou seja, a segunda coluna).

```python
print(df_dados_alunos.iloc[:, 1])
```

*Exercício 5: Utilizando a função loc para Selecionar Linhas e Colunas por Rótulo*

- Utilize o DataFrame "dados_alunos" criado no exercício 1.
- Use a função loc para selecionar e imprimir a idade do aluno com índice "C".

```python
print(df_dados_alunos.loc["C", "Idade"])
```

*Exercício 6: Utilizando a função iloc para Selecionar Linhas e Colunas por Índice Numérico*

Utilize o DataFrame "dados_alunos" criado no exercício 1.
Use a função iloc para selecionar e imprimir a idade do aluno na primeira linha e segunda coluna.

```python
print(df_dados_alunos.iloc[0, 1])
```
