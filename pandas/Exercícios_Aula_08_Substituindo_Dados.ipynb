# **Aula 08 - Substituindo Dados**

<br>
<br>

___


```python
import pandas as pd

dados = {
    "Nome": ["Maria", "João", "Pedro", "Ana"],
    "Idade": [25, 30, 35, 40]
}

dados_alunos = pd.DataFrame(dados)
```

*Exercício 1: Utilizando a função replace para Substituir Valores*

- Utilize o DataFrame "dados_alunos" criado
- Use a função replace para substituir a idade "30" por "31" no DataFrame.
- Imprima o DataFrame após a substituição.

```python
dados_alunos["Idade"] = dados_alunos["Idade"].replace(30, 31)
```

*Exercício 2: Utilizando a função replace com Dicionário de Substituição*

- Utilize o DataFrame "dados_alunos" criado no exercício 1.
- Crie um dicionário de substituição onde a idade "25" é substituída por "26" e a idade "35" é substituída por "36".
- Use a função replace com o dicionário de substituição para atualizar as idades.
- Imprima o DataFrame após a atualização.

```python
dados_alunos["Idade"] = dados_alunos["Idade"].replace({25: 26, 35: 36}, inplace=True)
```
