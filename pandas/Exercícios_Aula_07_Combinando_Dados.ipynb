*Exercício 1: Utilizando a função concat para Juntar DataFrames Verticalmente*

- Importe a biblioteca pandas.
- Crie dois DataFrames, "dados1" e "dados2", com colunas "Nome" e "Idade".
- Preencha os DataFrames com dados fictícios.
Utilize a função concat para juntar os DataFrames verticalmente (um abaixo do outro) e imprima o resultado.

```python
import pandas as pd
```

```python
d1= {"Nome": ["Ana", "Bruno"],"Idade": [20, 22]}

d2 = {"Nome": ["Carla", "Daniel"], "Idade": [19, 21]}

dados1 = pd.DataFrame(d1)
dados2 = pd.DataFrame(d2)

concat= pd.concat([dados1,dados2], ignore_index=True)

concat.head(4)
```

*Exercício 2: Utilizando a função concat para Juntar DataFrames Horizontalmente*

- Utilize os DataFrames "dados1" e "dados2" criados no exercício anterior.
- Use a função concat para juntar os DataFrames horizontalmente (um ao lado do outro) e imprima o resultado.

```python
horizontal = pd.concat([dados1, dados2], axis=1)

horizontal.head(4)
```

*Exercício 3: Utilizando a função merge para Mesclar DataFrames por Coluna Comum*

- Crie dois DataFrames, "vendedores" e "vendas", com informações sobre vendedores e suas vendas, respectivamente.
- Use a função merge para mesclar os DataFrames com base na coluna comum "ID" e imprima o resultado.

```python
dado_vendedor ={"ID": [1, 2, 3],"Nome": ["Ana", "Bruno", "Carla"]}

dado_vendas = {"ID": [1, 2, 1, 3],"Produto": ["Notebook", "Mouse", "Teclado", "Monitor"],"Valor": [3000, 150, 250, 1200]}

vendedor = pd.DataFrame(dado_vendedor)
vendas = pd.DataFrame(dado_vendas)

merge = pd.merge(vendedor,vendas, on="ID")

merge.head(6)
```

*Exercício 4: Utilizando a função merge com Mescla Interna e Colunas Diferentes*

- Utilize os DataFrames "vendedores" e "vendas" criados no exercício anterior.
- Use a função merge para mesclar os DataFrames com base em colunas diferentes ("ID" em "vendedores" e "Vendedor_ID" em "vendas").
- Imprima o resultado da mescla.

```python
vendas = vendas.rename(columns={'ID': 'Vendedor_ID'})

merge_colunas = pd.merge(vendedor,vendas,left_on="ID", right_on="Vendedor_ID")

merge_colunas.head()
```

*Exercício 5: Utilizando a função join para Unir DataFrames por Índice*

- Importe a biblioteca pandas.
- Crie dois DataFrames, "alunos" e "notas", com informações sobre alunos e suas notas, respectivamente.
- Use a função set_index para definir o índice de ambos os DataFrames como a coluna "ID".
- Utilize a função join para unir os DataFrames por índice e imprima o resultado.

```python
d1_5 = {"ID": [1, 2, 3], "Nome": ["Lucas", "Marina", "Pedro"]}

d2_5 = {"ID": [1, 2, 3],"Nota": [8.5, 9.0, 7.5]}
```

```python
alunos = pd.DataFrame(d1_5)
notas = pd.DataFrame(d2_5)

alunos = alunos.set_index("ID")
notas = notas.set_index("ID")

resultado_join = alunos.join(notas)

print(resultado_join)
```
