*Exercício 1: Gráficos*

Carregue um CSV/EXCEL que você já utiliza no seu trabalho e crie um gráfico com pandas de cada tipo:
- barras
- dispersão
- histograma
- linha
- box


```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
read = pd.read_csv('Exercícios_Aula_04_2026_QS_World University_Rankings.csv')
```

```python
df = pd.DataFrame(read)
df.head()
```

```python
df.plot(kind='scatter', x='Country/Territory', y='International student  RANK')
```

```python
df.plot(kind='line', x='Country/Territory', y='International student  RANK', marker='o')
```

```python
df[['International student  RANK']].plot(kind='box')
```

```python
df.groupby('Status')['International student  RANK'].count().plot(kind='bar')
```
